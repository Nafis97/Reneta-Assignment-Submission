##Encoding Similar Data

```python
def data_encode(data, n):
  encode={}
  num = n
  for i in data:
    if i not in encode:
      encode[i]= num
      num+=1
  n+=10
  return encode, n
```

```python
Division={}
Gender= {}
Marital_status= {}
n= 0
Division, n= data_encode(df['Division'], n)
Gender, n = data_encode(df['Gender'], n)
Marital_status, n= data_encode(df['MaritalStatus'], n)
print(Division)
print(Gender)
print(Marital_status)
```

```python
def numerical_df (new_df, data, encode):
  for i in data:
    new_df = new_df.replace([i], encode[i] )
  return new_df
```

```python
new_df = df
new_df = numerical_df(new_df, df['Division'], Division)
new_df = numerical_df (new_df, df['Gender'], Gender)
new_df = numerical_df(new_df, df['MaritalStatus'], Marital_status)
new_df
```

```python
new_df = new_df.drop(['Customer Name'], axis=1)
new_df
```

```python
print('Accuracy score: {0:0.2f}'. format(correct_labels/float(target.size)))
```

##Clustering Data

```python
from sklearn.decomposition import PCA

# First: for every document we get its corresponding cluster
clusters = kmeans.predict(input)

# We train the PCA on the dense version of the tf-idf.
pca = PCA(n_components=3)
two_dim = pca.fit_transform(input)

scatter_x = two_dim[:, 0] # first principle component
scatter_y = two_dim[:, 1] # second principle component
plt.style.use('ggplot')

fig, ax = plt.subplots()
fig.set_size_inches(10,9)

# color map for NUMBER_OF_CLUSTERS we have
cmap = {0: 'green', 1: 'blue', 2: 'red',3: 'yellow',4: 'violet', 5:'purple', 6:'orange', 7:'gray', 8: 'maroon', 9:'olive'}
#cmap = {0: 'green', 1: 'blue', 2: 'red', 3:'yellow'}

# group by clusters and scatter plot every cluster
# with a colour and a label
for group in np.unique(clusters):
    ix = np.where(clusters == group)
    ax.scatter(scatter_x[ix], scatter_y[ix], c=cmap[group], label=group)

ax.legend()
plt.xlabel("PCA 0")
plt.ylabel("PCA 1")
plt.show()
```

##Hyper-parameter Tuning

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import silhouette_score

clustering = KMeans(n_init=10)
param_grid = {
    "n_clusters": range(2, 9),
}


def scorer(estimator, X, y=None):
    y_pred = estimator.fit_predict(X)
    return silhouette_score(X, y_pred)


grid = GridSearchCV(clustering, param_grid=param_grid, cv=5, scoring=scorer)
grid.fit(input)
pd.DataFrame(grid.cv_results_)
```
